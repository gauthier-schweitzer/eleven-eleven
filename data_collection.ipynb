```python
import pandas as pd
import numpy as np
import json
import requests
from pandas.io.json import json_normalize
```

```python
adress = 'https://www.rescuetime.com/anapi/data?key=B63aWNpfuCKI7K22MnCTZN0CjuDGSY9pYyA2v_Rl&perspective=interval&format=json&resolution_time=hour&restrict_kind=activity&restrict_begin=20018-07-19&restrict_end=2018-12-31'

response = requests.get(adress)
response
```

```python
data = pd.DataFrame(response.json()["rows"], columns=response.json()['row_headers'])
```

```python
data.to_csv("data.csv", sep=',')
```

```python
from datetime import datetime
data['Date']=data['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%dT%H:%M:%S'))
data['Hour']=data['Date'].apply(lambda x: x.hour)
data['Date_short']=data['Date'].apply(lambda x: x.date())
```

```python
data.head()
```

We create a group_by to check how many seconds are monitored each hour of the day, on average

```python
data_gb = data.groupby(by=["Hour","Date_short"])['Time Spent (seconds)'].sum().groupby("Hour").mean()
data_gb
```
